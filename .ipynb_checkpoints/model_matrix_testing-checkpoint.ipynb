```python
import import_ipynb
from turkish_toxicity_classification import classify, load_dataset, create_matrix
```

```python
X, y, df = load_dataset(path  = 'datasets/turkish_dataset//turkce_cumleler_kokler_corrected_50k.csv', rows = 42000)
print("\nVeri yuklendi!")
```

```python
# Modeli eğit ve sınıflandır
log_bow_y_test, log_bow_y_pred, log_bow_classifier = classify(X, y, modelType="LR", vectorizerType="BOW",save=True)
#Modeller: SVM, LOG, MNB, DT, KNN, RF
#Vektörleştirmeler: BOW, TF

if log_bow_y_test is not None and log_bow_y_pred is not None:
    create_matrix(log_bow_y_test, log_bow_y_pred, log_bow_classifier)   # Confusion Matrix'i çiz
```
